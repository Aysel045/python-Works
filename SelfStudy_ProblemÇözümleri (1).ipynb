```python
# Write a Python code that find the students who have maximum and minimum average at below.
grades={'Student-1': {'Lesson-1': 57, 'Lesson-2': 46, 'Lesson-3': 58, 'Lesson-4': 81, 'Lesson-5': 85}, 
               'Student-2': {'Lesson-1': 85, 'Lesson-2': 56, 'Lesson-3': 51, 'Lesson-4': 69, 'Lesson-5': 67}, 
               'Student-3': {'Lesson-1': 68, 'Lesson-2': 76, 'Lesson-3': 87, 'Lesson-4': 57, 'Lesson-5': 56}, 
               'Student-4': {'Lesson-1': 78, 'Lesson-2': 93, 'Lesson-3': 88, 'Lesson-4': 38, 'Lesson-5': 54}, 
               'Student-5': {'Lesson-1': 50, 'Lesson-2': 46, 'Lesson-3': 78, 'Lesson-4': 81, 'Lesson-5': 75}}
# Önce her bir öğrencinin ortalamasını bulalım.
avrg_st1 = (grades['Student-1']["Lesson-1"] + grades['Student-1']["Lesson-2"] + grades['Student-1']["Lesson-3"] + grades['Student-1']["Lesson-4"] + grades['Student-1']["Lesson-5"]) /5
avrg_st2 = (grades['Student-2']["Lesson-1"] + grades['Student-2']["Lesson-2"] + grades['Student-2']["Lesson-3"] + grades['Student-2']["Lesson-4"] + grades['Student-2']["Lesson-5"]) /5
avrg_st3 = (grades['Student-3']["Lesson-1"] + grades['Student-3']["Lesson-2"] + grades['Student-3']["Lesson-3"] + grades['Student-3']["Lesson-4"] + grades['Student-3']["Lesson-5"]) /5
avrg_st4 = (grades['Student-4']["Lesson-1"] + grades['Student-4']["Lesson-2"] + grades['Student-4']["Lesson-3"] + grades['Student-4']["Lesson-4"] + grades['Student-4']["Lesson-5"]) /5
avrg_st5 = (grades['Student-5']["Lesson-1"] + grades['Student-5']["Lesson-2"] + grades['Student-5']["Lesson-3"] + grades['Student-5']["Lesson-4"] + grades['Student-5']["Lesson-5"]) /5
# Ortalama notları bir dictionary de key lerle eşleştirelim.
avrg_tuple = {}
avrg_tuple["Student-1"] = avrg_st1
avrg_tuple["Student-2"] = avrg_st2
avrg_tuple["Student-3"] = avrg_st3
avrg_tuple["Student-4"] = avrg_st4
avrg_tuple["Student-5"] = avrg_st5
avrg_lst = avrg_tuple.values()  # value ları listeye çevirdik.
max_key = max(avrg_tuple, key=avrg_tuple.get)  # max. ortalamaya sahip öğrencinin key ini bulduk.
min_key = min(avrg_tuple, key=avrg_tuple.get)  # min ortalamaya sahip öğrencinin key ini bulduk.
print("Student has max. average  is ", max_key, ":", max(avrg_lst))
print("Student has min. average  is ", min_key, ":", min(avrg_lst))
```
